```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def compute_partial_dependence(model, X, feature, grid_resolution=500):
    """
    Compute 1D partial dependence for a single feature.
    
    Parameters:
        model: trained model with predict_proba()
        X: DataFrame of input features
        feature: name of feature to analyze
        grid_resolution: number of values to evaluate

    Returns:
        grid: the grid of feature values
        pdp: average predicted probability at each grid value
    """
    
    X_temp = X.copy()
    grid = np.linspace(X[feature].min(), X[feature].max(), grid_resolution)
    pdp = []

    for val in grid:
        X_temp[feature] = val
        preds = model.predict_proba(X_temp)[:, 1]
        pdp.append(np.mean(preds))

    return grid, pdp

feature_name = 'loan_percent_income'
grid, pdp_vals = compute_partial_dependence(rf, X_test, feature_name)

plt.figure(figsize=(8, 6))
plt.plot(grid, pdp_vals, label=f'PDP of {feature_name}')
plt.xlabel(feature_name)
plt.ylabel('Predicted Probability')
plt.title(f'Partial Dependence Plot for {feature_name}')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, Optional, Union
import pandas as pd

def compute_2d_pdp(model, X: pd.DataFrame, feature1: str, feature2: str, 
                   grid_resolution: int = 20, model_type: str = 'classifier') -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Compute 2D partial dependence plot for two features.
    
    Parameters:
    -----------
    model : trained model with predict or predict_proba method
    X : pandas DataFrame with feature data
    feature1, feature2 : str, names of features to analyze
    grid_resolution : int, number of grid points for each feature
    model_type : str, 'classifier' or 'regressor'
    
    Returns:
    --------
    grid1, grid2 : 1D arrays of grid values
    pdp : 2D array of partial dependence values
    """

    if not isinstance(X, pd.DataFrame):
        raise ValueError("X must be a pandas DataFrame")
    
    if feature1 not in X.columns or feature2 not in X.columns:
        raise ValueError(f"Features {feature1} and/or {feature2} not found in DataFrame")
    
    if len(X) == 0:
        raise ValueError("X cannot be empty")
    
    grid1 = np.linspace(X[feature1].quantile(0.01), X[feature1].quantile(0.99), grid_resolution)
    grid2 = np.linspace(X[feature2].quantile(0.01), X[feature2].quantile(0.99), grid_resolution)
    
    pdp = np.zeros((grid_resolution, grid_resolution))
    
    X_temp = X.copy()
    
    for i, val1 in enumerate(grid1):
        for j, val2 in enumerate(grid2):
            X_temp[feature1] = val1
            X_temp[feature2] = val2
            
            if model_type == 'classifier':
                if hasattr(model, 'predict_proba'):
                    preds = model.predict_proba(X_temp)

                    if preds.shape[1] == 2:
                        preds = preds[:, 1]
                    else:
                        preds = np.max(preds, axis=1)
                else:
                    preds = model.predict(X_temp)
            else:
                preds = model.predict(X_temp)
            
            pdp[i, j] = np.mean(preds)
    
    return grid1, grid2, pdp

def plot_2d_pdp(grid1: np.ndarray, grid2: np.ndarray, pdp: np.ndarray, 
                feature1: str, feature2: str, 
                title: Optional[str] = None, 
                figsize: Tuple[int, int] = (10, 8),
                cmap: str = 'viridis',
                levels: int = 20) -> None:
    """
    Create a 2D partial dependence plot.
    
    Parameters:
    -----------
    grid1, grid2 : 1D arrays of grid values
    pdp : 2D array of partial dependence values
    feature1, feature2 : str, feature names
    title : str, optional plot title
    figsize : tuple, figure size
    cmap : str, colormap name
    levels : int, number of contour levels
    """
    
    G1, G2 = np.meshgrid(grid1, grid2)
    
    plt.figure(figsize=figsize)
    
    contour = plt.contourf(G1, G2, pdp.T, cmap=cmap, levels=levels)
    
    plt.contour(G1, G2, pdp.T, levels=levels//2, colors='white', alpha=0.3, linewidths=0.5)
    
    cbar = plt.colorbar(contour, shrink=0.8)
    cbar.set_label('Predicted Probability/Value', rotation=270, labelpad=20)
    
    plt.xlabel(feature1.replace('_', ' ').title(), fontsize=12)
    plt.ylabel(feature2.replace('_', ' ').title(), fontsize=12)
    
    if title is None:
        title = f'2D Partial Dependence Plot: {feature1} vs {feature2}'
    plt.title(title, fontsize=14, pad=20)
    
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def create_2d_pdp_analysis(model, X_test: pd.DataFrame, feature1: str, feature2: str, 
                          grid_resolution: int = 20, model_type: str = 'classifier'):
    """
    Complete 2D PDP analysis with error handling.
    """
    try:
        g1, g2, pdp_matrix = compute_2d_pdp(model, X_test, feature1, feature2, 
                                           grid_resolution=grid_resolution, 
                                           model_type=model_type)
        
        plot_2d_pdp(g1, g2, pdp_matrix, feature1, feature2)
        
        print(f"PDP Analysis Summary:")
        print(f"Feature 1 ({feature1}): {g1.min():.3f} to {g1.max():.3f}")
        print(f"Feature 2 ({feature2}): {g2.min():.3f} to {g2.max():.3f}")
        print(f"PDP range: {pdp_matrix.min():.3f} to {pdp_matrix.max():.3f}")
        print(f"PDP mean: {pdp_matrix.mean():.3f}")
        
        return g1, g2, pdp_matrix
        
    except Exception as e:
        print(f"Error in PDP analysis: {str(e)}")
        return None, None, None

f1, f2 = 'person_age', 'loan_amnt'
g1, g2, pdp_matrix = create_2d_pdp_analysis(rf, X_test, f1, f2, grid_resolution=25)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from typing import Tuple, Optional, Union
import pandas as pd

def compute_2d_ale(model, X: pd.DataFrame, feature1: str, feature2: str, 
                   grid_resolution: int = 20, model_type: str = 'classifier') -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """
    Compute 2D Accumulated Local Effects (ALE) for two features.
    
    Parameters:
    -----------
    model : trained model with predict or predict_proba method
    X : pandas DataFrame with feature data
    feature1, feature2 : str, names of features to analyze
    grid_resolution : int, number of grid points for each feature
    model_type : str, 'classifier' or 'regressor'
    
    Returns:
    --------
    grid1, grid2 : 1D arrays of grid values (bin centers)
    ale : 2D array of accumulated local effects
    """

    if not isinstance(X, pd.DataFrame):
        raise ValueError("X must be a pandas DataFrame")
    
    if feature1 not in X.columns or feature2 not in X.columns:
        raise ValueError(f"Features {feature1} and/or {feature2} not found in DataFrame")
    
    if len(X) == 0:
        raise ValueError("X cannot be empty")
    
    # Create quantile-based bins for better distribution
    quantiles1 = np.linspace(0, 1, grid_resolution + 1)
    quantiles2 = np.linspace(0, 1, grid_resolution + 1)
    
    bins1 = X[feature1].quantile(quantiles1).values
    bins2 = X[feature2].quantile(quantiles2).values
    
    # Ensure unique bins
    bins1 = np.unique(bins1)
    bins2 = np.unique(bins2)
    
    # Adjust grid resolution if we have fewer unique values
    actual_res1 = len(bins1) - 1
    actual_res2 = len(bins2) - 1
    
    # Grid centers for plotting
    grid1 = (bins1[:-1] + bins1[1:]) / 2
    grid2 = (bins2[:-1] + bins2[1:]) / 2
    
    # Initialize ALE matrix
    ale = np.zeros((actual_res1, actual_res2))
    
    # Compute local effects for each bin
    for i in range(actual_res1):
        for j in range(actual_res2):
            # Find data points in current bin
            mask = ((X[feature1] >= bins1[i]) & (X[feature1] < bins1[i+1]) & 
                   (X[feature2] >= bins2[j]) & (X[feature2] < bins2[j+1]))
            
            if not mask.any():
                continue  # Skip empty bins
            
            X_bin = X[mask].copy()
            
            if len(X_bin) == 0:
                continue
            
            # Compute local effects by comparing predictions at bin edges
            effects = []
            
            # Create copies for perturbation
            X_low1_low2 = X_bin.copy()
            X_low1_high2 = X_bin.copy()
            X_high1_low2 = X_bin.copy()
            X_high1_high2 = X_bin.copy()
            
            # Set feature values to bin edges
            X_low1_low2[feature1] = bins1[i]
            X_low1_low2[feature2] = bins2[j]
            
            X_low1_high2[feature1] = bins1[i]
            X_low1_high2[feature2] = bins2[j+1]
            
            X_high1_low2[feature1] = bins1[i+1]
            X_high1_low2[feature2] = bins2[j]
            
            X_high1_high2[feature1] = bins1[i+1]
            X_high1_high2[feature2] = bins2[j+1]
            
            # Get predictions
            pred_low1_low2 = _get_predictions(model, X_low1_low2, model_type)
            pred_low1_high2 = _get_predictions(model, X_low1_high2, model_type)
            pred_high1_low2 = _get_predictions(model, X_high1_low2, model_type)
            pred_high1_high2 = _get_predictions(model, X_high1_high2, model_type)
            
            # Compute 2D local effect (second-order difference)
            local_effect = (pred_high1_high2 - pred_high1_low2 - pred_low1_high2 + pred_low1_low2)
            ale[i, j] = np.mean(local_effect)
    
    # Accumulate effects (integrate over the grid)
    ale_accumulated = np.zeros_like(ale)
    
    # Accumulate along first dimension
    for i in range(actual_res1):
        for j in range(actual_res2):
            ale_accumulated[i, j] = np.sum(ale[:i+1, j])
    
    # Accumulate along second dimension
    for i in range(actual_res1):
        for j in range(actual_res2):
            ale_accumulated[i, j] += np.sum(ale[i, :j+1])
    
    # Center the ALE values (subtract mean)
    ale_accumulated -= np.mean(ale_accumulated)
    
    return grid1, grid2, ale_accumulated

def _get_predictions(model, X: pd.DataFrame, model_type: str) -> np.ndarray:
    """Helper function to get predictions from model."""
    if model_type == 'classifier':
        if hasattr(model, 'predict_proba'):
            preds = model.predict_proba(X)
            if preds.shape[1] == 2:
                preds = preds[:, 1]  # Binary classification - take positive class
            else:
                preds = np.max(preds, axis=1)  # Multi-class - take max probability
        else:
            preds = model.predict(X)
    else:
        preds = model.predict(X)
    
    return preds

def plot_2d_ale(grid1: np.ndarray, grid2: np.ndarray, ale: np.ndarray, 
                feature1: str, feature2: str, 
                title: Optional[str] = None, 
                figsize: Tuple[int, int] = (10, 8),
                cmap: str = 'RdBu_r',
                levels: int = 20) -> None:
    """
    Create a 2D Accumulated Local Effects plot.
    
    Parameters:
    -----------
    grid1, grid2 : 1D arrays of grid values
    ale : 2D array of accumulated local effects
    feature1, feature2 : str, feature names
    title : str, optional plot title
    figsize : tuple, figure size
    cmap : str, colormap name (RdBu_r is good for centered data)
    levels : int, number of contour levels
    """
    
    G1, G2 = np.meshgrid(grid1, grid2)
    
    plt.figure(figsize=figsize)
    
    # Use symmetric colormap limits around zero
    vmax = np.max(np.abs(ale))
    vmin = -vmax
    
    contour = plt.contourf(G1, G2, ale.T, cmap=cmap, levels=levels, 
                          vmin=vmin, vmax=vmax)
    
    # Add contour lines
    plt.contour(G1, G2, ale.T, levels=levels//2, colors='black', 
               alpha=0.4, linewidths=0.5)
    
    # Add zero contour line
    plt.contour(G1, G2, ale.T, levels=[0], colors='black', 
               linewidths=1.5, alpha=0.8)
    
    cbar = plt.colorbar(contour, shrink=0.8)
    cbar.set_label('Accumulated Local Effect', rotation=270, labelpad=20)
    
    plt.xlabel(feature1.replace('_', ' ').title(), fontsize=12)
    plt.ylabel(feature2.replace('_', ' ').title(), fontsize=12)
    
    if title is None:
        title = f'2D Accumulated Local Effects: {feature1} vs {feature2}'
    plt.title(title, fontsize=14, pad=20)
    
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def create_2d_ale_analysis(model, X_test: pd.DataFrame, feature1: str, feature2: str, 
                          grid_resolution: int = 20, model_type: str = 'classifier'):
    """
    Complete 2D ALE analysis with error handling.
    """
    try:
        g1, g2, ale_matrix = compute_2d_ale(model, X_test, feature1, feature2, 
                                           grid_resolution=grid_resolution, 
                                           model_type=model_type)
        
        plot_2d_ale(g1, g2, ale_matrix, feature1, feature2)
        
        print(f"ALE Analysis Summary:")
        print(f"Feature 1 ({feature1}): {g1.min():.3f} to {g1.max():.3f}")
        print(f"Feature 2 ({feature2}): {g2.min():.3f} to {g2.max():.3f}")
        print(f"ALE range: {ale_matrix.min():.3f} to {ale_matrix.max():.3f}")
        print(f"ALE mean: {ale_matrix.mean():.3f} (should be close to 0)")
        print(f"ALE std: {ale_matrix.std():.3f}")
        
        return g1, g2, ale_matrix
        
    except Exception as e:
        print(f"Error in ALE analysis: {str(e)}")
        return None, None, None

# Usage example (same as original)
f1, f2 = 'person_age', 'loan_amnt'
g1, g2, ale_matrix = create_2d_ale_analysis(rf, X_test, f1, f2, grid_resolution=25)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Optional, Union, List
from sklearn.inspection import partial_dependence
import warnings

def compute_ice_curves(model, X: pd.DataFrame, feature: str, 
                      grid_resolution: int = 50, 
                      sample_size: Optional[int] = None,
                      model_type: str = 'classifier',
                      percentile_range: Tuple[float, float] = (0.01, 0.99)) -> Tuple[np.ndarray, np.ndarray]:
    """
    Compute Individual Conditional Expectation (ICE) curves for a feature.
    
    Parameters:
    -----------
    model : trained model with predict or predict_proba method
    X : pandas DataFrame with feature data
    feature : str, name of feature to analyze
    grid_resolution : int, number of grid points
    sample_size : int, optional, number of instances to sample (None = use all)
    model_type : str, 'classifier' or 'regressor'
    percentile_range : tuple, (min_percentile, max_percentile) for feature range
    
    Returns:
    --------
    feature_values : 1D array of feature values
    ice_curves : 2D array of shape (n_samples, n_grid_points)
    """
    # Input validation
    if feature not in X.columns:
        raise ValueError(f"Feature '{feature}' not found in DataFrame")
    
    # Sample data if requested
    if sample_size is not None and sample_size < len(X):
        X_sample = X.sample(n=sample_size, random_state=42)
    else:
        X_sample = X.copy()
    
    # Create feature value grid
    feature_min = X[feature].quantile(percentile_range[0])
    feature_max = X[feature].quantile(percentile_range[1])
    feature_values = np.linspace(feature_min, feature_max, grid_resolution)
    
    # Initialize ICE curves matrix
    ice_curves = np.zeros((len(X_sample), grid_resolution))
    
    # Compute ICE curves
    for i, feature_val in enumerate(feature_values):
        X_temp = X_sample.copy()
        X_temp[feature] = feature_val
        
        if model_type == 'classifier':
            if hasattr(model, 'predict_proba'):
                preds = model.predict_proba(X_temp)
                if preds.shape[1] == 2:
                    preds = preds[:, 1]  # Binary classification
                else:
                    preds = np.max(preds, axis=1)  # Multiclass
            else:
                preds = model.predict(X_temp)
        else:  # regressor
            preds = model.predict(X_temp)
        
        ice_curves[:, i] = preds
    
    return feature_values, ice_curves

def plot_ice_curves(feature_values: np.ndarray, ice_curves: np.ndarray, 
                   feature_name: str, X: pd.DataFrame = None,
                   color_by: Optional[str] = None,
                   show_pdp: bool = True,
                   alpha: float = 0.3,
                   sample_curves: Optional[int] = None,
                   figsize: Tuple[int, int] = (10, 6),
                   title: Optional[str] = None) -> None:
    """
    Plot ICE curves with optional PDP overlay and color coding.
    
    Parameters:
    -----------
    feature_values : 1D array of feature values
    ice_curves : 2D array of ICE curves
    feature_name : str, name of the feature
    X : pandas DataFrame, original data (needed for color_by)
    color_by : str, optional, feature name to color curves by
    show_pdp : bool, whether to show PDP as thick line
    alpha : float, transparency of ICE curves
    sample_curves : int, optional, number of curves to display
    figsize : tuple, figure size
    title : str, optional plot title
    """
    plt.figure(figsize=figsize)
    
    # Sample curves if requested
    if sample_curves is not None and sample_curves < ice_curves.shape[0]:
        indices = np.random.choice(ice_curves.shape[0], sample_curves, replace=False)
        ice_curves_plot = ice_curves[indices]
        X_plot = X.iloc[indices] if X is not None else None
    else:
        ice_curves_plot = ice_curves
        X_plot = X
    
    # Plot ICE curves
    if color_by is not None and X_plot is not None:
        # Color curves by another feature
        colors = X_plot[color_by].values
        
        # Normalize colors for colormap
        colors_norm = (colors - colors.min()) / (colors.max() - colors.min())
        
        # Plot each curve with color
        for i in range(ice_curves_plot.shape[0]):
            plt.plot(feature_values, ice_curves_plot[i], 
                    alpha=alpha, color=plt.cm.viridis(colors_norm[i]), linewidth=1)
        
        # Add colorbar
        sm = plt.cm.ScalarMappable(cmap='viridis', 
                                 norm=plt.Normalize(vmin=colors.min(), vmax=colors.max()))
        sm.set_array([])
        cbar = plt.colorbar(sm)
        cbar.set_label(color_by.replace('_', ' ').title(), rotation=270, labelpad=20)
        
    else:
        # Plot all curves in single color
        for i in range(ice_curves_plot.shape[0]):
            plt.plot(feature_values, ice_curves_plot[i], 
                    alpha=alpha, color='lightblue', linewidth=1)
    
    # Add PDP line
    if show_pdp:
        pdp_values = np.mean(ice_curves_plot, axis=0)
        plt.plot(feature_values, pdp_values, 
                color='red', linewidth=3, label='PDP (Average)', alpha=0.8)
        plt.legend()
    
    # Formatting
    plt.xlabel(feature_name.replace('_', ' ').title(), fontsize=12)
    plt.ylabel('Predicted Probability/Value', fontsize=12)
    
    if title is None:
        title = f'ICE Plot for {feature_name}'
        if color_by:
            title += f' (colored by {color_by})'
    plt.title(title, fontsize=14, pad=20)
    
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_ice_with_interactions(model, X: pd.DataFrame, target_feature: str, 
                             interaction_feature: str,
                             grid_resolution: int = 50,
                             sample_size: int = 100,
                             model_type: str = 'classifier',
                             figsize: Tuple[int, int] = (12, 8)) -> None:
    """
    Create ICE plot specifically designed to show feature interactions.
    
    Parameters:
    -----------
    model : trained model
    X : pandas DataFrame with feature data
    target_feature : str, feature to vary (x-axis)
    interaction_feature : str, feature to color by (shows interaction)
    grid_resolution : int, number of grid points
    sample_size : int, number of instances to sample
    model_type : str, 'classifier' or 'regressor'
    figsize : tuple, figure size
    """
    # Compute ICE curves
    feature_values, ice_curves = compute_ice_curves(
        model, X, target_feature, grid_resolution, sample_size, model_type
    )
    
    # Create subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    # Plot 1: ICE curves colored by interaction feature
    plt.sca(ax1)
    plot_ice_curves(feature_values, ice_curves, target_feature, X.iloc[:sample_size], 
                   color_by=interaction_feature, show_pdp=True, alpha=0.4)
    
    # Plot 2: ICE curves grouped by interaction feature quantiles
    plt.sca(ax2)
    interaction_values = X[interaction_feature].iloc[:sample_size]
    
    # Create quantile groups
    quantiles = [0.25, 0.5, 0.75]
    q_values = interaction_values.quantile(quantiles)
    
    colors = ['blue', 'green', 'orange', 'red']
    labels = ['Low', 'Medium-Low', 'Medium-High', 'High']
    
    # Group curves by quantiles
    for i, (q_low, q_high, color, label) in enumerate(zip(
        [interaction_values.min()] + q_values.tolist(),
        q_values.tolist() + [interaction_values.max()],
        colors, labels
    )):
        mask = (interaction_values >= q_low) & (interaction_values <= q_high)
        if mask.sum() > 0:
            curves_group = ice_curves[mask]
            
            # Plot individual curves
            for j in range(curves_group.shape[0]):
                plt.plot(feature_values, curves_group[j], 
                        alpha=0.3, color=color, linewidth=1)
            
            # Plot group average
            plt.plot(feature_values, np.mean(curves_group, axis=0), 
                    color=color, linewidth=3, label=f'{label} {interaction_feature}')
    
    plt.xlabel(target_feature.replace('_', ' ').title(), fontsize=12)
    plt.ylabel('Predicted Probability/Value', fontsize=12)
    plt.title(f'ICE Plot: {target_feature} by {interaction_feature} Groups', fontsize=14)
    plt.legend()
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

def create_ice_analysis(model, X: pd.DataFrame, feature: str, 
                       interaction_feature: Optional[str] = None,
                       sample_size: int = 200,
                       model_type: str = 'classifier') -> None:
    """
    Complete ICE analysis with multiple visualization options.
    
    Parameters:
    -----------
    model : trained model
    X : pandas DataFrame
    feature : str, primary feature to analyze
    interaction_feature : str, optional, feature to show interactions with
    sample_size : int, number of instances to sample
    model_type : str, 'classifier' or 'regressor'
    """
    print(f"=== ICE Analysis for {feature} ===")
    print(f"Sample size: {min(sample_size, len(X))}")
    print(f"Feature range: {X[feature].min():.3f} to {X[feature].max():.3f}")
    
    try:
        # Basic ICE plot
        feature_values, ice_curves = compute_ice_curves(
            model, X, feature, sample_size=sample_size, model_type=model_type
        )
        
        if interaction_feature is None:
            # Simple ICE plot
            plot_ice_curves(feature_values, ice_curves, feature, show_pdp=True)
        else:
            # ICE plot with interaction analysis
            plot_ice_with_interactions(model, X, feature, interaction_feature, 
                                     sample_size=sample_size, model_type=model_type)
        
        # Print summary statistics
        print(f"\nICE Curve Statistics:")
        print(f"Number of curves: {ice_curves.shape[0]}")
        print(f"Prediction range: {ice_curves.min():.3f} to {ice_curves.max():.3f}")
        print(f"Average prediction: {ice_curves.mean():.3f}")
        print(f"Std deviation: {ice_curves.std():.3f}")
        
        # Measure interaction strength (variance of slopes)
        slopes = np.diff(ice_curves, axis=1)
        slope_variance = np.var(slopes, axis=0).mean()
        print(f"Interaction strength (slope variance): {slope_variance:.6f}")
        
    except Exception as e:
        print(f"Error in ICE analysis: {str(e)}")

# Example usage:
# Basic ICE plot
create_ice_analysis(rf, X_test, 'person_age', sample_size=100)

# ICE plot with interaction analysis
create_ice_analysis(rf, X_test, 'person_age', 
                   interaction_feature='loan_percent_income', 
                   sample_size=100)

# Manual plotting with more control
feature_values, ice_curves = compute_ice_curves(rf, X_test, 'loan_amnt', sample_size=50)
plot_ice_curves(feature_values, ice_curves, 'loan_amnt', X_test.iloc[:50], 
               color_by='loan_percent_income', show_pdp=True, alpha=0.5)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Tuple, Optional, List, Dict, Any
import warnings
warnings.filterwarnings('ignore')

# Note: You'll need to install shap first: pip install shap
import shap

class SHAPInteractionAnalyzer:
    """
    A comprehensive class for analyzing feature interactions using SHAP.
    """
    
    def __init__(self, model, X_train: pd.DataFrame, model_type: str = 'tree'):
        """
        Initialize SHAP analyzer.
        
        Parameters:
        -----------
        model : trained model
        X_train : pandas DataFrame, training data for background
        model_type : str, 'tree', 'linear', 'kernel', or 'deep'
        """
        self.model = model
        self.X_train = X_train
        self.model_type = model_type
        self.explainer = None
        self.shap_values = None
        self.shap_interaction_values = None
        self.feature_names = X_train.columns.tolist()
        
        # Initialize explainer based on model type
        self._initialize_explainer()
    
    def _initialize_explainer(self):
        """Initialize appropriate SHAP explainer based on model type."""
        if self.model_type == 'tree':
            # For tree-based models (RandomForest, XGBoost, LightGBM)
            self.explainer = shap.TreeExplainer(self.model)
        elif self.model_type == 'linear':
            # For linear models
            self.explainer = shap.LinearExplainer(self.model, self.X_train)
        elif self.model_type == 'kernel':
            # For any model (slower but universal)
            background = shap.sample(self.X_train, 100)  # Use subset for speed
            self.explainer = shap.KernelExplainer(self.model.predict, background)
        elif self.model_type == 'deep':
            # For deep learning models
            background = self.X_train.iloc[:100]  # Use subset for speed
            self.explainer = shap.DeepExplainer(self.model, background)
        else:
            raise ValueError(f"Unsupported model_type: {self.model_type}")
    
    def compute_shap_values(self, X_test: pd.DataFrame, 
                          compute_interactions: bool = True) -> None:
        """
        Compute SHAP values and interaction values.
        
        Parameters:
        -----------
        X_test : pandas DataFrame, test data
        compute_interactions : bool, whether to compute interaction values
        """
        print("Computing SHAP values...")
        
        # Compute main SHAP values
        self.shap_values = self.explainer.shap_values(X_test)
        
        # Handle multi-class output
        if isinstance(self.shap_values, list):
            # Multi-class classification - use class 1 for binary, or specify class
            if len(self.shap_values) == 2:
                self.shap_values = self.shap_values[1]  # Binary classification
            else:
                self.shap_values = self.shap_values[0]  # Default to first class
        
        # Compute interaction values if requested and supported
        if compute_interactions and hasattr(self.explainer, 'shap_interaction_values'):
            print("Computing SHAP interaction values...")
            try:
                self.shap_interaction_values = self.explainer.shap_interaction_values(X_test)
                
                # Handle multi-class output for interactions
                if isinstance(self.shap_interaction_values, list):
                    if len(self.shap_interaction_values) == 2:
                        self.shap_interaction_values = self.shap_interaction_values[1]
                    else:
                        self.shap_interaction_values = self.shap_interaction_values[0]
                        
            except Exception as e:
                print(f"Could not compute interaction values: {e}")
                self.shap_interaction_values = None
        
        print("SHAP computation complete!")
    
    def plot_interaction_summary(self, X_test: pd.DataFrame, 
                               max_features: int = 10) -> None:
        """
        Plot summary of feature interactions.
        
        Parameters:
        -----------
        X_test : pandas DataFrame
        max_features : int, maximum number of features to show
        """
        if self.shap_interaction_values is None:
            print("Interaction values not computed. Run compute_shap_values first.")
            return
        
        # Calculate interaction strengths
        interaction_strengths = np.abs(self.shap_interaction_values).mean(axis=0)
        
        # Get upper triangle (avoid double counting)
        n_features = len(self.feature_names)
        interaction_matrix = np.zeros((n_features, n_features))
        
        for i in range(n_features):
            for j in range(i+1, n_features):
                interaction_matrix[i, j] = interaction_strengths[i, j]
                interaction_matrix[j, i] = interaction_strengths[i, j]
        
        # Create heatmap
        plt.figure(figsize=(12, 10))
        
        # Select top features by interaction strength
        feature_sums = np.sum(interaction_matrix, axis=1)
        top_indices = np.argsort(feature_sums)[-max_features:]
        
        selected_matrix = interaction_matrix[np.ix_(top_indices, top_indices)]
        selected_names = [self.feature_names[i] for i in top_indices]
        
        # Create heatmap
        sns.heatmap(selected_matrix, 
                   xticklabels=selected_names,
                   yticklabels=selected_names,
                   annot=True, 
                   fmt='.3f',
                   cmap='RdYlBu_r',
                   center=0)
        
        plt.title('Feature Interaction Strength Matrix', fontsize=16, pad=20)
        plt.xlabel('Features', fontsize=12)
        plt.ylabel('Features', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()
    
    def plot_dependence_with_interactions(self, X_test: pd.DataFrame, 
                                        feature: str, 
                                        interaction_feature: Optional[str] = None,
                                        figsize: Tuple[int, int] = (10, 6)) -> None:
        """
        Create SHAP dependence plot showing interactions.
        
        Parameters:
        -----------
        X_test : pandas DataFrame
        feature : str, primary feature to analyze
        interaction_feature : str, optional, feature to show interactions with
        figsize : tuple, figure size
        """
        if self.shap_values is None:
            print("SHAP values not computed. Run compute_shap_values first.")
            return
        
        plt.figure(figsize=figsize)
        
        # Auto-detect interaction feature if not provided
        if interaction_feature is None and self.shap_interaction_values is not None:
            feature_idx = self.feature_names.index(feature)
            interaction_strengths = np.abs(self.shap_interaction_values[:, feature_idx, :]).mean(axis=0)
            interaction_strengths[feature_idx] = 0  # Exclude self-interaction
            interaction_feature_idx = np.argmax(interaction_strengths)
            interaction_feature = self.feature_names[interaction_feature_idx]
        
        # Create dependence plot
        shap.dependence_plot(
            feature, 
            self.shap_values, 
            X_test,
            interaction_index=interaction_feature,
            show=False
        )
        
        plt.title(f'SHAP Dependence Plot: {feature}' + 
                 (f' (colored by {interaction_feature})' if interaction_feature else ''))
        plt.tight_layout()
        plt.show()
    
    def plot_interaction_scatter(self, X_test: pd.DataFrame, 
                               feature1: str, feature2: str,
                               figsize: Tuple[int, int] = (10, 6)) -> None:
        """
        Create scatter plot of interaction effects between two features.
        
        Parameters:
        -----------
        X_test : pandas DataFrame
        feature1, feature2 : str, features to analyze
        figsize : tuple, figure size
        """
        if self.shap_interaction_values is None:
            print("Interaction values not computed. Run compute_shap_values first.")
            return
        
        idx1 = self.feature_names.index(feature1)
        idx2 = self.feature_names.index(feature2)
        
        # Get interaction values
        interaction_values = self.shap_interaction_values[:, idx1, idx2]
        
        # Create subplots
        fig, axes = plt.subplots(1, 2, figsize=figsize)
        
        # Plot 1: Feature1 vs Feature2, colored by interaction strength
        scatter = axes[0].scatter(X_test[feature1], X_test[feature2], 
                                c=interaction_values, cmap='RdYlBu_r', alpha=0.7)
        axes[0].set_xlabel(feature1.replace('_', ' ').title())
        axes[0].set_ylabel(feature2.replace('_', ' ').title())
        axes[0].set_title(f'Feature Values\n(colored by interaction strength)')
        plt.colorbar(scatter, ax=axes[0], label='Interaction Value')
        
        # Plot 2: Interaction strength distribution
        axes[1].hist(interaction_values, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
        axes[1].set_xlabel('Interaction Value')
        axes[1].set_ylabel('Frequency')
        axes[1].set_title('Distribution of Interaction Values')
        axes[1].axvline(0, color='red', linestyle='--', alpha=0.7, label='No Interaction')
        axes[1].legend()
        
        plt.suptitle(f'SHAP Interaction Analysis: {feature1}  {feature2}', 
                    fontsize=14, y=1.02)
        plt.tight_layout()
        plt.show()
        
        # Print summary
        print(f"Interaction Summary for {feature1}  {feature2}:")
        print(f"Mean interaction value: {interaction_values.mean():.4f}")
        print(f"Std interaction value: {interaction_values.std():.4f}")
        print(f"Max interaction value: {interaction_values.max():.4f}")
        print(f"Min interaction value: {interaction_values.min():.4f}")
    
    def find_top_interactions(self, top_k: int = 10) -> List[Tuple[str, str, float]]:
        """
        Find the top k feature interactions by strength.
        
        Parameters:
        -----------
        top_k : int, number of top interactions to return
        
        Returns:
        --------
        List of tuples: (feature1, feature2, interaction_strength)
        """
        if self.shap_interaction_values is None:
            print("Interaction values not computed. Run compute_shap_values first.")
            return []
        
        # Calculate interaction strengths
        interaction_strengths = np.abs(self.shap_interaction_values).mean(axis=0)
        
        # Get upper triangle pairs
        interactions = []
        n_features = len(self.feature_names)
        
        for i in range(n_features):
            for j in range(i+1, n_features):
                strength = interaction_strengths[i, j]
                interactions.append((self.feature_names[i], self.feature_names[j], strength))
        
        # Sort by strength
        interactions.sort(key=lambda x: x[2], reverse=True)
        
        return interactions[:top_k]
    
    def create_interaction_report(self, X_test: pd.DataFrame, 
                                top_k: int = 5) -> None:
        """
        Create a comprehensive interaction analysis report.
        
        Parameters:
        -----------
        X_test : pandas DataFrame
        top_k : int, number of top interactions to analyze
        """
        print("=== SHAP Interaction Analysis Report ===\n")
        
        # Find top interactions
        top_interactions = self.find_top_interactions(top_k)
        
        if not top_interactions:
            print("No interaction data available.")
            return
        
        print(f"Top {top_k} Feature Interactions:")
        print("-" * 50)
        for i, (feat1, feat2, strength) in enumerate(top_interactions, 1):
            print(f"{i}. {feat1}  {feat2}: {strength:.4f}")
        
        print(f"\nDetailed Analysis of Top 3 Interactions:")
        print("=" * 50)
        
        # Analyze top 3 interactions in detail
        for i, (feat1, feat2, strength) in enumerate(top_interactions[:3], 1):
            print(f"\n{i}. {feat1}  {feat2} (Strength: {strength:.4f})")
            print("-" * 40)
            
            # Create visualizations
            self.plot_interaction_scatter(X_test, feat1, feat2)
            self.plot_dependence_with_interactions(X_test, feat1, feat2)

# Convenience functions for quick analysis
def quick_shap_interaction_analysis(model, X_train: pd.DataFrame, 
                                  X_test: pd.DataFrame, 
                                  model_type: str = 'tree',
                                  sample_size: int = 500) -> SHAPInteractionAnalyzer:
    """
    Quick SHAP interaction analysis with default settings.
    
    Parameters:
    -----------
    model : trained model
    X_train : pandas DataFrame, training data
    X_test : pandas DataFrame, test data
    model_type : str, type of model ('tree', 'linear', 'kernel', 'deep')
    sample_size : int, number of test samples to analyze
    
    Returns:
    --------
    SHAPInteractionAnalyzer instance
    """
    # Sample data for faster computation
    if len(X_test) > sample_size:
        X_test_sample = X_test.sample(n=sample_size, random_state=42)
    else:
        X_test_sample = X_test
    
    # Initialize analyzer
    analyzer = SHAPInteractionAnalyzer(model, X_train, model_type)
    
    # Compute SHAP values
    analyzer.compute_shap_values(X_test_sample, compute_interactions=True)
    
    # Create summary plots
    analyzer.plot_interaction_summary(X_test_sample)
    
    # Create detailed report
    analyzer.create_interaction_report(X_test_sample)
    
    return analyzer

# Quick analysis
analyzer = quick_shap_interaction_analysis(rf, X_train, X_test, 
                                          model_type='tree', sample_size=300)

# Manual analysis with more control
analyzer = SHAPInteractionAnalyzer(rf, X_train, model_type='tree')
analyzer.compute_shap_values(X_test.sample(200), compute_interactions=True)

# Individual plots
analyzer.plot_interaction_summary(X_test.sample(200))
analyzer.plot_dependence_with_interactions(X_test.sample(200), 'loan_amnt')
analyzer.plot_interaction_scatter(X_test.sample(200), 'loan_amnt', 'loan_percent_income')

# Find top interactions
top_interactions = analyzer.find_top_interactions(top_k=10)
print(top_interactions)
```
